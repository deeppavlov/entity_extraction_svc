```python
import bs4
import html2text
import inscriptis
import libextract.api
import requests
import trafilatura

url = "https://www.theguardian.com/politics/live/2022/jul/01/boris-johnson-conservatives-chris-pincher-uk-politics-latest-live"
raw_html = requests.get(url).content

def save_and_print(path, result):
    with open(path, "w", encoding="utf-8") as save_f:
        save_f.write(result)
    print(result)
```

## BeautifulSoup
[docs](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#installing-a-parser)

```python
soup_native = bs4.BeautifulSoup(raw_html, "html.parser")
save_and_print("bs4_native.txt", soup_native.get_text(separator="\n", strip=True))
```

```python
soup_lxml = bs4.BeautifulSoup(raw_html, "lxml")
save_and_print("bs4_lxml.txt", soup_lxml.get_text(separator="\n", strip=True))
```

## html2text
[github](https://github.com/Alir3z4/html2text)

```python
save_and_print("html2text.txt", html2text.html2text(raw_html.decode("utf-8")))
```

## Libextract (very old)
[github](https://github.com/datalib/libextract)

```python
save_and_print(
    "libextract.txt",
    "\n---\n".join(node.text_content() for node in libextract.api.extract(raw_html))
)
```

## Inscriptis
[github](https://github.com/weblyzard/inscriptis)

```python
save_and_print("inscriptis.txt", inscriptis.get_text(raw_html.decode("utf-8")))
```

## Trafilatura (most used)
[github](https://github.com/adbar/trafilatura)

```python
save_and_print("trafilatura_all_defaults.txt", trafilatura.extract(raw_html))
```

```python
save_and_print("trafilatura_comments_false.txt", trafilatura.extract(raw_html, include_comments=False))
```
